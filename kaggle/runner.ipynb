## 前置条件
1. 数据集已上传到 Kaggle Dataset
2. GitHub Token 已准备好（下面会让你填）


```python
# ========== 1. 配置 ==========
import os
from pathlib import Path

# GitHub 配置
GITHUB_REPO = "git@github.com:KyleCream/EEGTokenizer.git"
GITHUB_BRANCH = "main"

# Kaggle 数据集路径
DATASET_PATH = "/kaggle/input/eeg-dataset"

# ========== 老板，请在这里填你的 GitHub Token ==========
GITHUB_TOKEN = ""  # 👈 老板，请在这里填你的 GitHub Token！！

# 工作目录
WORK_DIR = Path("/kaggle/working/EEGTokenizer")

print(f"工作目录: {WORK_DIR}")
print(f"数据集路径: {DATASET_PATH}")
```

```python
# ========== 2. 从 GitHub Pull 最新代码 ==========
# 确保使用HTTPS格式的URL而不是SSH格式
GITHUB_REPO = GITHUB_REPO.replace('git@github.com:', 'https://github.com/')

# 自动接受GitHub的主机密钥
!mkdir -p ~/.ssh
!ssh-keyscan github.com >> ~/.ssh/known_hosts 2>/dev/null || true

# 配置 git 避免交互式确认
!git config --global user.name "Kaggle Bot"
!git config --global user.email "kaggle@example.com"
!git config --global pull.rebase false

# 执行克隆
!cd /kaggle/working && git clone {GITHUB_REPO} || true
!cd {WORK_DIR} && git checkout {GITHUB_BRANCH} && git pull origin {GITHUB_BRANCH}
```

```python
# ========== 3. 安装依赖 ==========
!cd {WORK_DIR} && pip install -q -r requirements.txt
```

```python
# ========== 4. 运行实验 ==========
!cd {WORK_DIR} && python kaggle/run_experiment.py
```

```python
# ========== 5. Push 结果回 GitHub ==========
# 修改远程仓库URL，将token嵌入其中
!cd {WORK_DIR} && git remote set-url origin https://{GITHUB_TOKEN}@github.com/KyleCream/EEGTokenizer.git

# 现在push应该可以正常工作
!cd {WORK_DIR} && python kaggle/push_results.py
```
